```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
